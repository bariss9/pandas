#### 7- groupby() belirli bir stuna göre gruplama yapan fonskiyondur

```python
df = pd.DataFrame({
    "Kategori": ["A", "B", "A", "B", "A"],
    "Değer": [10, 20, 30, 40, 50]
})
print(df.groupby("Kategori").sum())
```
